# Encoding

```python
data = data.drop(['case', 'country'], axis=1)
data
```

```python
data.columns
```

```python
cc3_dummies = pd.get_dummies(data['cc3'])
data = pd.concat([data, cc3_dummies], axis=1)
data = data.drop('cc3', axis=1)
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
label_encoder=LabelEncoder()
data["banking_crisis"]=label_encoder.fit_transform(data["banking_crisis"])
Bank_mapping={index:label for index,label in enumerate(label_encoder.classes_)}
Bank_mapping
```

# Splitting

```python
from sklearn.model_selection import train_test_split
y = data['banking_crisis']
X = data.drop('banking_crisis', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
```

# Scaling

```python
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
X
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression 
logistic_regression= LogisticRegression() 
logistic_regression.fit(X_train,y_train) 
```

```python
y_pred=logistic_regression.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred) 
print(f"Test accuracy: {accuracy:.4f}")
```

```python
from sklearn.metrics import roc_auc_score
acc = roc_auc_score(y_test, y_pred)
print("Test set auc: {:.4f}".format(acc))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor()
dt.fit(X_train,y_train)
```

```python
y_pred = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
y_pred = dt.predict(X_test)
accuracy = accuracy_score(y_test, y_pred) 
print(f"Test accuracy: {accuracy:.4f}")
```

```python
from sklearn.metrics import roc_auc_score
acc = roc_auc_score(y_test, y_pred)
print("Test set auc: {:.4f}".format(acc))
```
